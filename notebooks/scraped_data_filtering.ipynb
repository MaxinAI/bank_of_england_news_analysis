```python
import pandas as pd
import pickle
pd.options.display.max_colwidth = 10000
import matplotlib.pyplot as plt
```

```python
!mv ../../named-entity-recognition/data/nnbm.pkl nnmb.pkl
```

```python
with open('nnmb.pkl','rb') as f:
    data = pickle.load(f)
```

```python
data = pd.DataFrame.from_dict(data)
```

```python
def filter_newlines(text):
    return ' '.join([x.strip() for x in text.split()])
```

```python
def list_to_text(text_list):
    return ' '.join(text_list)
```

```python
def filter_content(content):
    text = list_to_text(content)
    filtered_text = filter_newlines(text)
    return filtered_text
```

```python
filter_content(data.content[0])
```

```python
data.content = data.content.apply(filter_content)
```

```python
data.header = data.header.apply(filter_content)
```

```python
data = data.drop_duplicates('url')
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(14,7))
data.header.apply(lambda x: len(x)).plot(ax=ax1)
data.header.apply(lambda x: len(x)).hist(bins=20,ax=ax2)
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(14,7))
data.content.apply(lambda x: len(x)).plot(ax=ax1)
data.content.apply(lambda x: len(x)).hist(bins=20,ax=ax2)
```

```python
data.to_csv('bank_of_england_news.csv')
```
